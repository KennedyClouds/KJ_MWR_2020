Description: This code produces a two panel plot displaying RAP soundings. Portions of this code are adapted from examples provided by SHARPpy documentation at:
        https://sharppy.github.io/SHARPpy/auto_examples/plot_sounding.html

```python
import warnings # Silence the warnings from SHARPpy
warnings.filterwarnings("ignore")
import sharppy
import sharppy.plot.skew as skew
from sharppy.sharptab import utils
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.ticker import ScalarFormatter, MultipleLocator, FixedLocator
from sharppy.io.buf_decoder import BufDecoder

#
sounding_12utc = 'data/RAP/rap_kgfk_12utc.buf'
sounding_18utc = 'data/RAP/rap_kgfk_18utc.buf'

# Read in the sounding data and generate a Profile objects
dec = BufDecoder(sounding_12utc)
profs = dec.getProfiles() 
stn_id = dec.getStnId()
prof_12UTC = profs._profs[""][0] #get initial analysis

dec = BufDecoder(sounding_18utc)
profs = dec.getProfiles() 
stn_id = dec.getStnId()
prof_18UTC = profs._profs[""][0] #get initial analysis

pb_plot=1000
pt_plot=100
dp_plot=10
plevs_plot = np.arange(pb_plot,pt_plot-1,-dp_plot)
# Open up the text file with the data in columns (e.g. the sample OAX file distributed with SHARPpy)
titlea = '12 UTC RAP Analysis - KGFK - 24 February 2019'
titleb = '18 UTC RAP Analysis - KGFK - 24 February 2019'
# Set up the figure in matplotlib. Usegridspec for positioning
fig = plt.figure(figsize=(10, 12))
gs = gridspec.GridSpec(2,2,width_ratios=[7,1])
#-------------Create first SKEW-T---------------------------
ax = plt.subplot(gs[0, 0], projection='skewx')
ax.text(0.55, 1.05,titlea,horizontalalignment='center',fontsize=16,transform=ax.transAxes,weight='bold')
ax.text(0.01, 1.05,'(a)',horizontalalignment='center',fontsize=16,transform=ax.transAxes,weight='bold')
ax.grid(True)
plt.grid(True)

# Plot the background variables
presvals = np.arange(1000, 0, -10)

ax.semilogy(prof_12UTC.tmpc, prof_12UTC.pres, 'r', lw=2)
ax.semilogy(prof_12UTC.dwpc, prof_12UTC.pres, 'g', lw=2)
# Highlight the 0 C and -20 C isotherms.
l = ax.axvline(0, color='b', ls='--')
l = ax.axvline(-20, color='b', ls='--')

# Disables the log-formatting that comes with semilogy
ax.yaxis.set_major_formatter(ScalarFormatter())
ax.set_yticks(np.linspace(100,1000,10))
ax.set_ylim(1000,300)

# Draw the wind barbs axis and everything that comes with it.
ax.xaxis.set_major_locator(FixedLocator([-50,-40,-30,-20,-10,0,10]))
ax.set_xlim(-50,10)
ax2 = plt.subplot(gs[0,1:])
skew.plot_wind_axes(ax2)
u, v = utils.vec2comp(prof_12UTC.wdir,prof_12UTC.wspd)
skew.plot_wind_barbs(ax2, prof_12UTC.pres[1::2], u[1::2], v[1::2])

#-------------Create second SKEW-T---------------------------
ax = plt.subplot(gs[1:, 0], projection='skewx')
#skew.draw_title(ax, title,)
ax.text(0.55, 1.05,titleb,horizontalalignment='center',fontsize=16,transform=ax.transAxes,weight='bold')
ax.text(0.01, 1.05,'(b)',horizontalalignment='center',fontsize=16,transform=ax.transAxes,weight='bold')
ax.grid(True)
plt.grid(True)

# Plot the background variables
presvals = np.arange(1000, 0, -10)

ax.semilogy(prof_18UTC.tmpc, prof_18UTC.pres, 'r', lw=2)
ax.semilogy(prof_18UTC.dwpc, prof_18UTC.pres, 'g', lw=2)
# Highlight the 0 C and -20 C isotherms.
l = ax.axvline(0, color='b', ls='--')
l = ax.axvline(-20, color='b', ls='--')

# Disables the log-formatting that comes with semilogy
ax.yaxis.set_major_formatter(ScalarFormatter())
ax.set_yticks(np.linspace(100,1000,10))
ax.set_ylim(1000,300)

# Draw the wind barbs axis and everything that comes with it.
ax.xaxis.set_major_locator(FixedLocator([-50,-40,-30,-20,-10,0,10]))
ax.set_xlim(-50,10)
ax2 = plt.subplot(gs[1:,1:])
skew.plot_wind_axes(ax2)
u, v = utils.vec2comp(prof_18UTC.wdir,prof_18UTC.wspd)
skew.plot_wind_barbs(ax2, prof_18UTC.pres[1::2], u[1::2], v[1::2])



gs.update(left=0.05, bottom=0.05, top=0.95, right=1, wspace=0.025)
gs.tight_layout(fig)
plt.savefig('KJ_MWR_2020_Fig02_final.png',dpi=300,bbox_inches='tight')
plt.savefig('KJ_MWR_2020_Fig02_final.eps', format='eps')
plt.show()
```
